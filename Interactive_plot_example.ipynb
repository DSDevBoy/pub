```python
def f(h):
    plt.figure(figsize=(6,5))
    x = [0,10,20,0]
    y = [0,10,0,0]
    x = [0,10,20,0]
    y = [h+0,h+-10,h+0,h+0]


plt.close("all")

fig,_ = plt.subplots()
plt.subplots_adjust(left=0.25, bottom=0.25)

x = [0,10,20,0]
y = [0,10,0,0]
x2 = [0,10,20,0]
y2 = [10+0,10+(-10),10+0,10+0]
xi = [x,x2]
yi = [y,y2]

a0 = 10

l2, = plt.plot(x,y,linewidth=2, color='red')
l, = plt.plot(x2,y2,linewidth=2, color='blue')
#plt.show()

#plt.axis([0, 1, -10, 10])

axcolor = 'lightgoldenrodyellow'
axamp = plt.axes([0.25, 0.15, 0.65, 0.03], facecolor=axcolor)
samp = Slider(axamp, 'Dist', 0.0, 20.0, valinit=a0)

resetax = plt.axes([0.8, 0.025, 0.1, 0.04])
button = Button(resetax, 'Reset', color=axcolor, hovercolor='0.975')

rax = plt.axes([0.025, 0.5, 0.15, 0.15], facecolor=axcolor)
radio = RadioButtons(rax, ('red', 'blue', 'green'), active=1)
rax.text(0,1.1,"Shape 1")

rax2 = plt.axes([0.025, 0.25, 0.15, 0.15], facecolor=axcolor)
radio2 = RadioButtons(rax2, ('red', 'blue', 'green'), active=1)
rax2.text(0,1.1,"Shape 2")

# #axbox = plt.axes([0.025, 0.5, 0.01, 0.01])
# axbox = plt.axes([0.1, 0.05, 0.8, 0.075])
# text_box = TextBox(axbox, 'Evaluate', initial="Line 1")


def colorfunc(label):
    l.set_color(label)
    fig.canvas.draw_idle()
radio.on_clicked(colorfunc)

def colorfunc2(label):
    l2.set_color(label)
    fig.canvas.draw_idle()
radio2.on_clicked(colorfunc2)

def reset(event):
    y2 = np.array([[10,0,10,10]])
    l.set_ydata(y2)
    samp.reset()
    fig.canvas.draw_idle()
    #samp.reset()
button.on_clicked(reset)

    
def update(val):
    v = samp.val
    
    y2 = [v+0,v+(-10),v+0,v+0]
    x2 = [0,10,20,0]
    ymin = min(y2)
    ymax = max(y2)
    l.set_ydata(y2)
    l.set_xdata(x2)
    if ymax < 10:
        ymax = 10
    if ymin > 0:
        ymin = 0
    l.axes.set_ylim(ymin,ymax)
    #l.axes.set_xlim([ymin,ymax])
    fig.canvas.draw_idle()

samp.on_changed(update)
plt.show()
```
